```python
import pandas as pd
```

```python
def carga_de_dados(caminho):
  tabela = pd.read_csv(caminho, sep = ";", encoding="ISO-8859-1")
  return tabela

tabela_alunos = carga_de_dados("https://firebasestorage.googleapis.com/v0/b/todo-list-7eeed.appspot.com/o/tbl_alunos.csv?alt=media&token=59a7ed44-041a-473a-b259-d49e119f5964")
tabela_notas = carga_de_dados("https://firebasestorage.googleapis.com/v0/b/todo-list-7eeed.appspot.com/o/tblnotas.csv?alt=media&token=4f1334df-6ba4-456e-ae67-d2724a190011")
display(tabela_alunos)
display(tabela_notas)
```

```python
def juncao(df_alunos, df_notas):
  nota1 = df_notas.query("id_nota == 1")['nota'] 
  nota2 = df_notas.query("id_nota == 2")['nota']
  nota3 = df_notas.query("id_nota == 3")['nota']
  notas = {'Nota 1': nota1.values,
           'Nota 2': nota2.values,
           'Nota 3': nota3.values}
  tabelaNotasMedias = pd.DataFrame(notas)
  media = round(tabelaNotasMedias.mean(axis = 1), 2)
  tabelaNotasMedias['Media'] = media 
  tabelaNova = df_alunos.join(tabelaNotasMedias)
  return tabelaNova

tabelaCompleta = juncao(tabela_alunos, tabela_notas)
tabelaCompleta
```

```python
def gravar_tabela(df, string=None):
  csv = df.to_csv(string, index = False, encoding="ISO-8859-1", sep=';')
  return csv

tabela = gravar_tabela(tabelaCompleta)
print(tabela)
```

```python
def consulta(df, id = None, nome = None):
  if id is not None:
    infos = df.query(f'id_aluno == {id}')
    return infos
  elif nome is not None:
    infos = df.query(f'Nome == "{nome}"')
    return infos

infoEstudante = consulta(tabelaCompleta, 1410839469, "MATEUS ALAYO HIDALGO SCHULZ")
display(infoEstudante)
```

```python
def excluir(df, id):
  df.loc[df.id_aluno == id, 'Excluido'] = 1 
  return df

tabelaCompleta = excluir(tabelaCompleta, 1033221237)
tabelaCompleta = excluir(tabelaCompleta, 1033270786)
tabelaCompleta.head()
```

```python
def alteracao(df, id, novo_nome):
  df.loc[df.id_aluno == id, 'Nome'] = novo_nome
  return df

tabelaCompleta = alteracao(tabelaCompleta, 1033211931, "Maria da Silva")
tabelaCompleta = alteracao(tabelaCompleta, 1033242142, "Cassia Ferreira")
tabelaCompleta.head()
```

```python
def alteracaoNota(df, id, nota, novanota):
  df.loc[df.id_aluno == id, f'Nota {nota}'] = novanota
  df['Media'] = round(df.iloc[:, 3:6].mean(axis = 1), 2)
  return df

tabelaCompleta = alteracaoNota(tabelaCompleta, 1033211931, 2, 8)
tabelaCompleta.head()
```
